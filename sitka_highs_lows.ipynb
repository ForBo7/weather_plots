In this notebook, we will analyze and compare Sitka's maximum and minimum temperatures throught the year.

Let us first load the required libraries and file into a DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt

filename = 'data/sitka_weather_2018_simple.csv'
sitka_weather = pd.read_csv(filename)
sitka_weather.head()
```

We will now plot high highest and lowest temperatures of each day.

```python
# Convert the date column from type string to type datetime.
sitka_weather.loc[:, 'DATE'] = pd.to_datetime(sitka_weather.loc[:, 'DATE'],
                                              format='%Y-%m-%d')
# Plot the high temperatures by date.
plt.figure(dpi=227, figsize=(16, 10))
plt.style.use('seaborn')
sitka_weather.plot(x='DATE', y=['TMAX', 'TMIN'], color=['red', 'blue'],
                   alpha=0.5)
plt.fill_between(sitka_weather['DATE'], sitka_weather['TMAX'],
                 sitka_weather['TMIN'], facecolor='purple',
                 alpha=0.1)

# Format plot
plt.title("Daily high and low temperatures - 2018", fontsize=24)
plt.xlabel('', fontsize=16)
plt.ylabel('Temperature (F)', fontsize=16)
plt.tick_params(axis='both', which='both', labelsize=16)

# Display plot.
plt.show()
```

The minimum and maximum temperatures for each day increase during the months
and decrease during the winter months as expected. It is also seen that the
and minimum temperatures for each day stay in tandem; that is when the maximum
temperature increases, the minimum temperature also increases and vice versa.
